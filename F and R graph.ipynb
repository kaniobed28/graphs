```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_excel(r"C:\Users\kanio\Desktop\rfdata.xlsx")
```

```python
data.head()
```

```python
data.FREQUENCY.mean()
```

```python
F_mean = data.FREQUENCY.mean()
R_mean = data.RANGE.mean()
```

```python
F_total = data.FREQUENCY.sum()
R_total = data.RANGE.sum()
```

```python
OVER_ALL_TOTAL = F_total + R_total
```

```python
F_prob = F_total/OVER_ALL_TOTAL
R_prob = R_total/OVER_ALL_TOTAL
```

```python
plt.rcParams['figure.figsize'] =(36,10)
plt.rcParams['font.size'] = 22
plt.pie([F_mean,R_mean],colors=['red','black'])
plt.legend([f"FSF:{F_mean.round(2)}",f"RSF:{R_mean.round(2)}"],loc =1)


plt.savefig("pie.jpeg", dpi=500)
plt.show()
```

```python
plt.rcParams['figure.figsize'] =(36,10)
plt.rcParams['font.size'] = 22
plt.pie([F_prob,R_prob],colors=['red','black'])
plt.legend([f"FSF:{F_prob.round(2)}",f"RSF:{R_prob.round(2)}"],loc =1)


plt.savefig("F R graph.jpeg", dpi=500)
plt.show()
```
