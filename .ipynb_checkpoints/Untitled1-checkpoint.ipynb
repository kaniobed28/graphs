```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("Book1.xlsx")
```

```python
data = data.dropna(0)
```

```python
plt.bar(data["MONTHS"],data["range occurance"])
```

```python
fig, axs = plt.subplots(2, 1, figsize=(25, 20))
axs[0].bar(data["MONTHS"],data["range occurance"],color="black",label="RSF")
axs[1].bar(data["MONTHS"],data["Frequency occurance"],color="red",label="FSF")
axs[0].set_ylim([0,100])
axs[1].set_ylim([0,100])
axs[0].set_ylabel("% Occurance ", size=26, weight ='bold')
axs[1].set_ylabel("% Occurance ", size=26, weight ='bold')
axs[0].legend(fontsize = 20)
axs[1].legend(fontsize = 20)
axs[0].tick_params(direction = "out")
axs[0].secondary_yaxis("right")
axs[0].secondary_xaxis("top")

axs[1].tick_params(direction = "out")
axs[1].secondary_yaxis("right")
axs[1].secondary_xaxis("top")


plt.savefig("dataplot.png", dpi=500)# this line of code is responsible for saving the graph as an image(png file).
plt.show()
```
